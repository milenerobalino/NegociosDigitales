Umbral de cuotas

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\P5E003-O\Desktop\olist_order_payments_dataset.csv")

# Cálculo del umbral de cuotas (Percentil 75)
umbral_cuotas = df['payment_installments'].quantile(0.75)

print("Umbral de cuotas (Percentil 75):", umbral_cuotas)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(8,5))
sns.histplot(df['payment_installments'], bins=10, kde=True)

plt.axvline(
    umbral_cuotas,
    linestyle='--',
    label='Umbral de cuotas (P75)'
)

plt.title("Distribución del número de cuotas y umbral")
plt.xlabel("Número de cuotas")
plt.ylabel("Frecuencia")
plt.legend()

plt.show()
```

Cual es la moda de las personas que pagan en cuotas

```python
# Filtrar solo pagos en cuotas
pagos_en_cuotas = df[df['payment_installments'] > 1]

# Cálculo de la moda
moda_cuotas = pagos_en_cuotas['payment_installments'].mode()[0]

print("Moda de las personas que pagan en cuotas:", moda_cuotas)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    x='payment_installments',
    data=pagos_en_cuotas
)

plt.axvline(
    moda_cuotas,
    linestyle='--',
    label='Moda de cuotas'
)

plt.title("Moda del número de cuotas en pagos fraccionados")
plt.xlabel("Número de cuotas")
plt.ylabel("Cantidad de pagos")
plt.legend()

plt.show()
```
